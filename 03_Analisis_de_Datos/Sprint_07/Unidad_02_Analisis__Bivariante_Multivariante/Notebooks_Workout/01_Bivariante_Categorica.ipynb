#### Caso 1. Seguros: Bivariante response-state

Nos han pedido que estudiemos la respuesta a las campaas de marketing segn la demografa por eso un conjunto de analisis que tenemos que hacer s o s es el de la variable/columna `response` con todas las que consideremos demogrficas, por ejeplo: `state`, `education`,`marital_status`,`gender`,`eploymentstatus`. Nosotros vamos a hacer el anlisis de relacin con `state` como ejemplo.

**#1  Comparar sus frecuencias** tal como las obtuvimos en el univariante

```python
pinta_distribucion_categoricas(df_seguros,["response","state"], relativa= True, mostrar_valores= True)
```

As, antes de "combinarlas", lo que espero es que, si no hay interaccin o relacin entre ellas. En concreto, como "response" es la variable directora, es la que quiero ver en funcin de la otra, y por tanto lo que yo quiero ver es si hay variacin de la distribucin de "YES"/"NO" por estado. Por eso mira como lo muestro:

**#2 Haz la comparacin con la variable ms importante como directora** es decir que sea la de la "leyenda", en nuestro caso es ponerla en la funcin la segunda

```python
plot_categorical_relationship_fin(df_seguros, "state","response", show_values= True)
```

Parece que s, que las respuestas se distribuyen igual independientemente del estado, pero sera mejor verlo con algunos nmeros quizs, y adems no en frecuencia absoluta sino en relativa. Juguemos con nuestra funcin:


```python
plot_categorical_relationship_fin(df_seguros, "state","response", relative_freq = True, show_values = True)
```

Pues guau, est claro que no hay relacin entre una y otra (clava en cada estado la medida general). Esto ya s es un buen resultado para mi anlisis, si quiero mejorar la respuesta de mis campaas no hay un estado que me sirva de ejemplo, pero tambin me dice que no es algo que dependa del lugar de residencia de mis clientes. 

```python
# En nmeros, por si no te crees el grfico
df_seguros.groupby("state", as_index = False)["response"].value_counts(normalize = True)
```
