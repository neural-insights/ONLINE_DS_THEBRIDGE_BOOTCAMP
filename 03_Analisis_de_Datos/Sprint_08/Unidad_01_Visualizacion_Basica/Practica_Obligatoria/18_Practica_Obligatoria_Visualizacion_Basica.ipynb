### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# Top 5 ciudades más pobladas
top_cities = df.sort_values(by='population_total', ascending=False).head(5)

# Colores con transparencia
colors = ['blue', 'gray', 'olive', 'brown', 'purple']
alphas = [0.7] * 5

plt.figure(figsize=(8, 6))
bars = plt.bar(top_cities['city'], top_cities['population_total'], color=colors, alpha=0.7)

# Anotaciones de población centradas
for i, (city, pop) in enumerate(zip(top_cities['city'], top_cities['population_total'])):
    plt.text(i, pop + 20000, str(pop), ha='center', va='bottom', fontsize=12)

plt.title("Top 5 ciudades más pobladas")
plt.ylabel("Población total")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

# Guardar fuente original
original_font = mpl.rcParams['font.family']

# Fuente personalizada temporária
mpl.rcParams['font.family'] = ['Bradley Hand ITC']

plt.figure(figsize=(8, 6))
plt.hist(df['area_total_sq_mi'].dropna(), bins=50, color='steelblue', edgecolor='black')
plt.title("Histograma del área total ocupada", fontsize=18)
plt.xlabel("Área total (mi²)")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.tight_layout()
plt.show()

# Restaurar fonte original
mpl.rcParams['font.family'] = original_font
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['area_water_sq_mi'], df['population_total'], alpha=0.6, color='teal')
plt.title("Área de agua vs Población")
plt.xlabel("Área de agua (mi²)")
plt.ylabel("Población total")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(6, 6))
sns.boxplot(x=df['elevation_m'], color='coral')
plt.title("Boxplot de la elevación de las ciudades (m)")
plt.ylabel("Elevación (m)")
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 12))

# Subplot 1: Ciudades más pobladas
top_cities = df.sort_values(by='population_total', ascending=False).head(5)
colors = ['blue', 'gray', 'olive', 'brown', 'purple']
axs[0, 0].bar(top_cities['city'], top_cities['population_total'], color=colors, alpha=0.7)
for i, (city, pop) in enumerate(zip(top_cities['city'], top_cities['population_total'])):
    axs[0, 0].text(i, pop + 20000, str(pop), ha='center', va='bottom', fontsize=10)
axs[0, 0].set_title("Top 5 ciudades más pobladas")

# Subplot 2: Histograma
axs[0, 1].hist(df['area_total_sq_mi'].dropna(), bins=50, color='steelblue', edgecolor='black')
axs[0, 1].set_title("Histograma del área total")

# Subplot 3: Scatter área agua vs población
axs[1, 0].scatter(df['area_water_sq_mi'], df['population_total'], alpha=0.6, color='teal')
axs[1, 0].set_title("Área de agua vs Población")
axs[1, 0].set_xlabel("Área de agua (mi²)")
axs[1, 0].set_ylabel("Población total")

# Subplot 4: Boxplot elevación
sns.boxplot(x=df['elevation_m'], ax=axs[1, 1], color='coral')
axs[1, 1].set_title("Boxplot de la elevación")

plt.tight_layout()
plt.show()
```
